## Chess(King-Rook vs. King) Endgame Classification
For introduction and rules of Chess, see [Wiki page](https://en.wikipedia.org/wiki/Chess). 

<img src="chess.png" width="400">

We will use Chess(King-Rook vs. King) Data Set from UCI machine learning repository. (See introduction [here](https://archive.ics.uci.edu/ml/datasets/Chess+(King-Rook+vs.+King)). This database has 28056 possible instances of chess endgame situations where the white has a king and a rook and the black has only a king. The goal is to determine what is the minimum depth for the white to win.

The dataset has 6 attributes. Each of them can take 8 values, listed as following:

1. White King file (column a - h) 
2. White King rank (row 1 - 8) 
3. White Rook file 
4. White Rook rank 
5. Black King file 
6. Black King rank 

And the label/class is the least number of steps that white must use to win. (draw if more than 16). The following is how the data set looks like.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.utils import shuffle
chess = pd.read_csv('./krkopt_data.txt', header=None)
chess.columns = ['wkf', 'wkr', 'wrf', 'wrr', 'bkf', 'bkr', 'class']
chess = shuffle(chess, random_state = 0)
chess.head(10)
```

Next we convert these values into boolean features using the same one-hot encoding trick we described for TIC-TAC-TOE game. Deleting symmetric features for the white king and drop the first for the others, we get a data set with $36$ boolean features. 

Next we randomly pick $70\%$ of the data to  be our training set and the remaining for testing. Training set looks like the following:

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
d_wkf = pd.get_dummies(chess['wkf'], prefix='wkf')
d_wkr = pd.get_dummies(chess['wkr'], prefix='wkr')
d_wrf = pd.get_dummies(chess['wrf'], prefix='wrf', drop_first=True)
d_wrr = pd.get_dummies(chess['wrr'], prefix='wrr', drop_first=True)
d_bkf = pd.get_dummies(chess['bkf'], prefix='bkf', drop_first=True)
d_bkr = pd.get_dummies(chess['bkr'], prefix='bkr', drop_first=True)
chess_new = pd.concat([d_wkf, d_wkr, d_wrf, d_wrr, d_bkf, d_bkr, chess['class']], axis=1)
X = chess_new.iloc[:, :-1]
y = chess_new['class']
le = LabelEncoder()
y = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.head(10)
```

## 1. Bagging and Random Forest.
Recall that the classifier we get by bagging with decision trees as our base classifier is called Random Forest. Using the Bagging Meta Classifier implemented in 'sklearn.ensemble'. See [Scikit Learn Bagging](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.BaggingClassifier.html#sklearn.ensemble.BaggingClassifier) for how to use it. (Don't use the random forest classifier in 'sklearn.ensemble' directly since it has other parameters and you may get weird results). Using information gain as your splitting criterion and train a random forest with number of classifiers in the following set:

$$n = \{ 2^i | i = 0, 1, ..., 11\}$$

Plot the training and testing accruracy and justify your plot.

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import hamming_loss
n_max = 12
Err_Train = np.zeros(n_max)
Err_Test = np.zeros(n_max)
indices = 2**np.array(range(0,n_max))
#==========Write your code below=====
for i in range(0,n_max):
    clf  = BaggingClassifier(base_estimator=DecisionTreeClassifier(criterion='entropy'), 
                             n_estimators=indices[i]).fit(X_train, y_train)
    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)
    Err_Train[i] = hamming_loss(y_train, y_pred_train)
    Err_Test[i] = hamming_loss(y_test, y_pred_test)
#================================

plt.semilogx(indices,Err_Train, label = "training")
plt.semilogx(indices,Err_Test, label = "testing")
plt.legend()
```

### Justify your plot :


# <font color='blue'> Explanation for Problem 1 </font> #
## <font color='blue'> Training Error </font> ##
<font color='blue'> As expected, as we increase the number of estimators used in the ensemble, the error of the training set decreases. The plot starts to decline quickly from 0.2 after 2 base estimators. The ensemble reaches a training error of 0 after 30 base estimators and does not see any more improvements beyond this point. In addition, the training error is lower than the testing error, as the ensemble is expected to perform better on data it has already seen. </font>

## <font color='blue'> Testing Error </font> ##
<font color='blue'> The plot for the testing error is more jagged than the training error. In some cases, such as between 20 and 30 base estimators, the ensemble actually performs slightly worse due to the increase in the number of base estimators. However, overall, we see an improvement in performance as we increase the number of base estimators from 1 to 2^11, resulting in a testing error of 0.4 at 2^11 base estimators. As we are adding more trees to the forest, we expect to model have better generalizability and become less prone to overfitting. </font>

## 2. Adaboost.

Decision trees with small maximum depth won't give us good performance because of limited complexity. In this problem, we use adaboost algorithm to reduce the bias of the model and hopefully this will give us better performance. Using decision trees with maximum depth 10, 20, 50, 100 as you base classifiers, try [Adaboost algorithm](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostClassifier.html#sklearn.ensemble.AdaBoostClassifier) with number of iterations in the following set:

$$T = \{ 2^i | i = 0, 1, ..., 11\}$$

Plot your training and testing error and justify your plots. When do you get zero training error? Explain.

```python
from sklearn.ensemble import AdaBoostClassifier
n_max = 12
Err_Train = np.zeros(n_max)
Err_Test = np.zeros(n_max)
indices = 2**np.array(range(0,n_max))
depth = [10,20,50,100]
#==========Write your code below=====
#You can repeat this block multiple times for your experiments.
for j in range(0,4):
    for i in range(0,n_max):
        clf = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(criterion='entropy',max_depth=depth[j]), n_estimators=indices[i]).fit(X_train, y_train)
        y_pred_train = clf.predict(X_train)
        y_pred_test = clf.predict(X_test)
        Err_Train[i] = hamming_loss(y_train, y_pred_train)
        Err_Test[i] = hamming_loss(y_test, y_pred_test)
    plt.semilogx(indices,Err_Train, label = "training")
    plt.semilogx(indices,Err_Test, label = "testing")
    tit = "Max Depth = " + str(depth[j])
    plt.title(tit)
    plt.legend()
    plt.show()
```

### Justify and Explain

# <font color='blue'> Explanation for Problem 2 </font> #
<font color='blue'> -At a depth of 10, as we increase the number of base estimators, the training error and testing error both increase slightly then decrease significantly. The training error reaches a minimum of 0.1 near a base estimator count of 500. The testing error reaches a minimum of around 0.4. As expected, the training error is lower than the testing error. With low depth and high number of iterations, the Adaboost estimator is less likely to overfit on the training data. For the training error, we cannot reach an error of 0 at the given base estimator counts.</font>
    
<font color='blue'> -At a depth of 20, we see a rapid decline in training error from 0.1 to 0 as we increase the number of base estimators beyond 2. For testing error, we see an increase then a decrease in testing error from above 0.5 to around 0.35. This is likely the optimal configuration, as the training error reaches a minimum of zero at a base estimator count after 2, and testing error reaches a minimum of around 0.35. </font>
    
<font color='blue'> -At a depth of 50, we see a much different behavior for both the training and testing error. Both plots are almost completely flat. The training error is zero across all base estimator counts and the testing error is approximately 0.475. This is likely due to overfitting due to the high depth of the tree base classifier. With a depth of 50, the model is completely tuned to the training data and cannot generalize well on the testing data. Thus, we do not see any decline in testing error with increasing number of iterations. We can conclude that boosting works well for weaker base classifiers.</font>
    
<font color='blue'> -At a depth of 100, the argument is the same as that of a depth of 50. There is significant overfitting and cannot generalize on the testing data, resulting in a relatively high testing error.</font>

## 3. Boosting Complex Classifiers.

Using random forest with 10 trees and max_depth 50 as your base classifier, train an AdaBoost classifier with number 
of iterations from 

$$T = \{ 2^i | i = 0, 1, ..., 6\}$$

Plot the training and testing error. Justify your plot.

```python
from sklearn.ensemble import AdaBoostClassifier
n_max = 7
Err_Train = np.zeros(n_max)
Err_Test = np.zeros(n_max)
indices = 2**np.array(range(0,n_max))
#==========Write your code below=====
for i in range(0,n_max):
    randomforest = BaggingClassifier(base_estimator=DecisionTreeClassifier(criterion='entropy', max_depth=50),n_estimators=10)
    clf = AdaBoostClassifier(base_estimator=randomforest, n_estimators=indices[i]).fit(X_train, y_train)
    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)
    Err_Train[i] = hamming_loss(y_train, y_pred_train)
    Err_Test[i] = hamming_loss(y_test, y_pred_test)
#================================
plt.semilogx(indices,Err_Train, label = "training")
plt.semilogx(indices,Err_Test, label = "testing")
plt.legend()
```

### Justify

# <font color='blue'> Explanation for Problem 3 </font> #
## <font color='blue'> Training Error </font> ##
<font color='blue'> Initially, the training error is already quite low. At an iteration number of 2, the training error reaches and remains at zero. As expected, the training error is lower than the testing error, as the model is expected to perform better on data it has seen already. </font>

## <font color='blue'> Testing Error </font> ##
<font color='blue'>For the testing error, the error is quite high and only decreases slightly as we increase the number of iterations. The error decreases from around 0.45 to around 0.4 as we increase the number of iterations from 1 to 2. The likely reason for this high error is that the the base estimator is now a relatively strong classifier (random forest). The random forest is composed of 10 trees, each with a maximum depth of 50. The Adaboost algorithm works the best when the base estimator is weak, so by using an ensemble, the end result is a strong classifier. Using a strong classifier as base classifier is not ideal. When compared with Adaboost with a base classifier of a decision tree of depth 50 and 100, the Adaboost model with the random forest base classifier actually performs slightly better (when looking at testing error). </font>
