# Problem 1 Write your own Gaussian Naive Bayes code
Using the skeleton provided below, write your own code for learning and inference using Gaussian Naive Bayes model. You can use the skeleton provided in the second problem to verify whether you are writing it correctly. Reading through chapter 1.9.1 in the following link and the slide from last year can be useful:

````python
import math

def gnb_train(X, y, output_label):
    output_size = len(output_label)
    prior = np.zeros(output_size)
    mean = np.zeros((output_size, X.shape[1]))
    std = np.zeros((output_size, X.shape[1]))
# ======= Your Code Here =======
    summation = 0
    variance = 0
    classOccurence = 0
    classCount = 0
    for i in range(0,output_size):
        # iterate over the jth index of element k
        for j in range(0,len(X[0])):    
            for k in range(0,len(X)):
                if y[k] == output_label[i]:
                    summation+=X[k][j]
                    classOccurence+=1
            mean[i][j] = float(summation)/classOccurence
            for k in range(0,len(X)):
                if y[k] == output_label[i]:
                    variance += ((X[k][j]-float(summation)/classOccurence)**2)/classOccurence
            std[i][j] = (variance)**0.5
            summation = 0
            variance = 0
            classOccurence = 0
    # computing the prior
    for i in range(0,output_size):
        for j in range(0,len(y)):
            if y[j] == output_label[i]:
                classCount += 1
        prior[i] = classCount/len(y)
        classCount = 0
    return prior, mean, std

def gauss(x, mu, sigma):
    x = float(x - mu) / sigma
    return math.exp(-x*x/2.0) / (math.sqrt(2.0*math.pi) * sigma)

def gnb_predict(X, prior, mean, std, output_label):
    predict = []
    piproduct = 1
# ======= Your Code Here =======
    for i in range(0,len(X)):
        prob = [0]*len(output_label)
        for j in range(0,len(output_label)):
            priorp = prior[j]
            for k in range(0,len(X[0])):
                piproduct*=gauss(X[i][k],mean[j][k],std[j][k])
            prob[j] = priorp*piproduct
            piproduct = 1
        max_prob = max(prob)
        predict.append(output_label[prob.index(max_prob)])
        prob = [0]*len(output_label)
    return predict
  
````

# Problem 2: Inference on IRIS dataset
Using your own GNB functions, implementing Gaussian Naive Bayes algorithm for the first feature, the first two features, the first three features and the first four features. Output the error for each experiment and explain using the visualization of the dataset.

````python
from sklearn.metrics import hamming_loss
output_label = list(set(y_trn))
for i in range(1,5):
    prior, mean, std = gnb_train(X_trn[:,0:i], y_trn, output_label)
    y_pred = gnb_predict(X_tst[:,0:i], prior, mean, std, output_label)
    error = hamming_loss(y_tst, y_pred)
    print("Test error using first", i, "features:",  error)
````

### Problem 2 Explanation ###

Based on the visualization of the dataset, we can see that for certain features, there can be a lot of overlap between the distribution of two labels, while other features are more clearly separable. By only using one feature, we are risking the potential for that feature to take on similar values for different labels. By using multiple features, we are making sure that one feature will not skew the probabilities. In addition, it is unlikely that two different labels will have the same distribution across all four features. By chaining the four features together, we are getting maximal separation.</font>

````python
err = np.zeros(300)
X_trn_new = np.hstack((X_trn, np.tile(X_trn[:, [0]], 20)))
for i in range(1,300):
    X_trn_new = np.hstack((X_trn, np.tile(X_trn[:, [0]], i)))
    X_tst_new = np.hstack((X_tst, np.tile(X_tst[:, [0]], i)))
    prior, mean, std = gnb_train(X_trn_new, y_trn, output_label)
    y_pred = gnb_predict(X_tst_new, prior, mean, std, output_label)
    err[i] = hamming_loss(y_tst, y_pred)
plt.plot(err)
````

### Question 3 Explanation ###
<font color='blue'> As we repeat the first feature more and more times, the error increases to a value of around 35% at i = 175. By repeating the first feature, we are assigning that feature to a higher weight. Thus, if the first feature has a slightly low probability, by giving it a higher weight, the overall probability is lowered even further. Thus, we expect the error to grow due to the increased weight on the first feature. Perhaps increasing the feature space by adding new features (not repeated ones) might be better. If we repeat the first feature for infinitely many times, I believe the testing accuracy will not be zero.
